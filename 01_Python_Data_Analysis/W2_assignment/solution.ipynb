# 펭귄 문제 해결

펭귄 데이터셋으로 pandas 데이터 처리 실습을 해보자

```python
import pandas as pd
import numpy as np
```

# 1. 데이터 불러오기 및 확인
penguins.csv 파일을 불러와 df 변수에 저장하고, 상위 5개 행을 출력하여 데이터의 구조를 확인하세요.

```python
# 1. 데이터 불러오기
df = pd.read_csv('penguins.csv')
print("--- 1. 상위 5개 행 ---")
print(df.head())
```

# 2. 데이터 정보 확인
데이터프레임의 전체 행 수, 컬럼별 데이터 타입, Non-null(결측치가 아닌) 개수를 확인하는 함수를 실행하세요.

```python
# 2. 데이터 정보 확인
print("\n--- 2. 데이터 정보 ---")
print(df.info())
```

# 3. 결측치(Missing Value) 개수 파악
각 컬럼별로 결측치(NaN)가 총 몇 개씩 있는지 확인하세요.

```python
# 3. 결측치 개수 파악
print("\n--- 3. 컬럼별 결측치 수 ---")
print(df.isnull().sum())
```

# 4. 결측치 처리 (행 삭제)
결측치가 하나라도 포함된 행을 모두 제거하고, 그 결과를 df_clean이라는 새로운 변수에 저장하세요. (이후 문제는 df_clean을 사용합니다.)

```python
# 4. 결측치 처리 (제거)
df_clean = df.dropna()
print("\n--- 4. 결측치 제거 후 행 개수 ---")
print(f"원본: {len(df)}, 제거 후: {len(df_clean)}")
```

# 5. 조건 필터링 (단일 조건)
species가 'Adelie'인 펭귄들의 데이터만 필터링하여 출력하세요

```python
# 5. Adelie 종 필터링
adelie_penguins = df_clean[df_clean['species'] == 'Adelie']
print("\n--- 5. Adelie 종 데이터 (상위 3개) ---")
print(adelie_penguins.head(3))
```

# 6. 조건 필터링 (다중 조건)
island가 'Torgersen'이면서 body_mass_g이 4000g 이상인 펭귄 데이터를 필터링하세요.

```python
# 6. 다중 조건 필터링
target_penguins = df_clean[(df_clean['island'] == 'Torgersen') & (df_clean['body_mass_g'] >= 4000)]
print("\n--- 6. Torgersen 섬 & 4000g 이상 펭귄 ---")
print(target_penguins.head())
```

# 7. 기술 통계 확인
bill_length_mm (부리 길이) 컬럼의 평균, 표준편차, 최소값, 최대값 등 주요 통계량을 한 번에 확인하세요.

```python
# 7. 기술 통계
print("\n--- 7. 부리 길이 기술 통계 ---")
print(df_clean['bill_length_mm'].describe())
```

# 8. 정렬 (Sorting)
데이터를 body_mass_g (몸무게) 기준으로 내림차순(큰 값부터 작은 값 순서) 정렬하고, 상위 10개 행만 출력하세요.

```python
# 8. 정렬
sorted_df = df_clean.sort_values(by='body_mass_g', ascending=False)
print("\n--- 8. 몸무게 내림차순 정렬 (상위 5개) ---")
print(sorted_df[['species', 'body_mass_g']].head())
```

# 9. 파생 변수 생성 (Column 추가)
body_mass_g 컬럼은 g 단위입니다. 이를 kg 단위로 변환하여 body_mass_kg라는 새로운 컬럼을 생성하세요. (예: 3750.0 -> 3.75)

```python
# 9. 파생 변수 생성
df_clean['body_mass_kg'] = df_clean['body_mass_g'] / 1000
print("\n--- 9. kg 컬럼 생성 확인 ---")
print(df_clean[['body_mass_g', 'body_mass_kg']].head())
```

# 10. 그룹화 및 집계 (GroupBy)
species (종)별로 body_mass_g (몸무게)의 평균값을 구하세요.

```python
# 10. 종별 몸무게 평균
species_mean = df_clean.groupby('species')['body_mass_g'].mean()
print("\n--- 10. 종별 몸무게 평균 ---")
print(species_mean)
```

# 11. 고유값 개수 확인 (Value Counts)
island (섬) 컬럼에 어떤 섬들이 있는지, 그리고 각 섬에 펭귄이 몇 마리씩 있는지 개수를 세어보세요.

```python
# 11. 섬별 펭귄 수 확인
print("\n--- 11. 섬별 펭귄 수 ---")
print(df_clean['island'].value_counts())
```

# 12. 상관관계 분석 (Advanced)
flipper_length_mm (날개 길이)와 body_mass_g (몸무게) 간의 상관계수를 구하여 두 변수 간의 관계를 확인하세요.

```python
# 12. 상관관계
corr = df_clean['flipper_length_mm'].corr(df_clean['body_mass_g'])
print("\n--- 12. 날개 길이와 몸무게의 상관계수 ---")
print(round(corr, 4))
```

# 데이터 다시 불러오기(초기화)
셀을 실행하고 다음 문제를 계속 풀어주세요~

```python
df = pd.read_csv('penguins.csv')
```

# 13. 결측치 대체 (Imputation)
앞서 결측치를 제거했지만, 이번에는 body_mass_g 컬럼의 결측치를 해당 컬럼의 평균값으로 채워보세요. (fillna 사용)

```python
# 13. 결측치 평균으로 대체
# 평균값 계산
mean_mass = df['body_mass_g'].mean()
# 결측치 채우기
df['body_mass_g'] = df['body_mass_g'].fillna(mean_mass)
print("--- 13. 결측치 대체 후 확인 (isnull sum) ---")
print(df['body_mass_g'].isnull().sum())
```

# 14. 고유값 확인 (Unique Values)
species 컬럼에 어떤 종들이 있는지 중복 없이 리스트 형태로 확인하세요. 

```python
# 14. 고유값 확인
unique_species = df['species'].unique()
print("\n--- 14. Species 고유값 ---")
print(unique_species)
```

# 15. 행/열 선택 (loc/iloc)
데이터프레임에서 10번째부터 15번째 행까지, 그리고 species와 body_mass_g 컬럼만 선택하여 출력하세요.

```python
# 15. loc/iloc 행/열 선택
# iloc는 정수 인덱스 기반 (10:16은 10~15행), loc는 라벨 기반
# 여기서는 편의상 loc 사용 예시 (컬럼명 직접 지정 가능)
subset = df.loc[10:15, ['species', 'body_mass_g']]
print("\n--- 15. 특정 행/열 선택 ---")
print(subset)
```

# 16. 문자열 처리 (String Manipulation)
sex 컬럼의 값을 모두 대문자로 변경하세요. (예: 'Male' -> 'MALE')

```python
# 16. 문자열 대문자 변환
# 문자열 접근자 .str 사용
df['sex'] = df['sex'].str.upper()
print("\n--- 16. Sex 컬럼 대문자 변환 ---")
print(df['sex'].head())
```

# 17. 함수 적용 (apply)
body_mass_g가 4000g 이상이면 'Heavy', 아니면 'Light'라고 표시하는 새로운 컬럼 mass_class를 만드세요. (apply와 lambda 함수 사용)

```python
# 17. apply 함수 적용
# 4000 이상이면 Heavy, 아니면 Light
df['mass_class'] = df['body_mass_g'].apply(lambda x: 'Heavy' if x >= 4000 else 'Light')
print("\n--- 17. mass_class 생성 확인 ---")
print(df[['body_mass_g', 'mass_class']].head())
```

# 18. 피벗 테이블 (Pivot Table)
species와 sex를 기준으로 펭귄들의 body_mass_g 평균을 구하는 피벗 테이블을 만드세요. (행: species, 열: sex, 값: body_mass_g 평균)

```python
# 18. 피벗 테이블
pivot_df = df.pivot_table(index='species', columns='sex', values='body_mass_g', aggfunc='mean')
print("\n--- 18. 피벗 테이블 (종/성별 몸무게 평균) ---")
print(pivot_df)
```

# 19. 데이터 매핑 (Map)
sex 컬럼의 'Male'을 0으로, 'Female'을 1로 변환하는 코드를 작성하세요. (단, 변환 전 sex 컬럼의 결측치는 제거하거나 적절히 처리된 상태여야 합니다. 여기서는 결측치를 제외한 데이터로 실습하세요.)

```python
# 19. Map (매핑)
# 먼저 결측치가 없는 임시 데이터프레임으로 실습
df_map = df.dropna(subset=['sex']).copy()
mapping_dict = {'MALE': 0, 'FEMALE': 1} # 위에서 대문자로 바꿨으므로 키값도 대문자
df_map['sex_code'] = df_map['sex'].map(mapping_dict)
print("\n--- 19. Sex 매핑 결과 (상위 5개) ---")
print(df_map[['sex', 'sex_code']].head())
```

# 20. 데이터 샘플링 (Sampling)
전체 데이터에서 무작위로 10개의 행을 추출(sample)하세요.

```python
# 20. 샘플링
sample_df = df.sample(n=10, random_state=42) # random_state는 재현성을 위해 설정
print("\n--- 20. 무작위 샘플 10개 ---")
print(sample_df.index.tolist()) # 추출된 인덱스 확인
```

# 21. 컬럼 이름 변경 (Rename)
bill_length_mm 컬럼의 이름을 bill_len으로, bill_depth_mm 컬럼의 이름을 bill_depth로 변경하세요.

```python
# 21. 컬럼 이름 변경
df = df.rename(columns={'bill_length_mm': 'bill_len', 'bill_depth_mm': 'bill_depth'})
print("\n--- 21. 컬럼 이름 변경 확인 ---")
print(df.columns)
```

# 22. 데이터 저장 (To CSV)
모든 전처리가 끝난 데이터프레임을 processed_penguins.csv라는 이름의 파일로 저장하세요. (인덱스는 저장하지 않음: index=False)

```python
# 22. 데이터 저장
df.to_csv('processed_penguins.csv', index=False)
print("\n--- 22. 저장 완료 ---")
```
