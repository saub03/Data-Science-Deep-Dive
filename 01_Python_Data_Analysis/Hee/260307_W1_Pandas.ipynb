## 1. 자료구조

### 1.1. Series

왼쪽에는 index 오른쪽에는 해당 index의 값 출력

```python
obj = pd.Series([4, 7, -5, 3])
obj
```

```python
# 배열과 색인 객체
print(obj.array)
print(obj.index)
```

```python
obj2 = pd.Series([4, 7, -5, 3], index=["d", "b", "a", "c"]) # index 지정
obj2["a"]
```

```python
obj2["d"] = 6 # 색인으로 접근
obj2[["a", "b", "d"]] # 색인의 배열으로 접근
```

```python
obj2[obj2 < 0]
```

```python
obj2 * 2
```

```python
np.exp(obj2)
```

```python
dict_data = {'A':1, 'B':2, 'C':3, 'D':4}
obj3 = pd.Series(dict_data)
obj3
```

```python
obj3.to_dict()
```

```python
index = ['B', 'C', 'D', 'E']
obj4=pd.Series(dict_data, index=index) # 인덱스대로 정렬
obj4
```

```python
print(obj4.isna())
print(obj4.notna())
```

```python
obj3 + obj4
```

```python
obj3.name = "nums" # Series 객체와 색인은 이름을 가진다.
obj3.index.name = "state"
print(obj3)
```

```python
obj.index = ["A", "B", "C", "D"] # 대입으로 인덱스를 변경할 수 있다.
obj
```

## 1.2. DataFrame

```python
data = {"state":["Ohio", "Ohio", "Ohio", "Nevada", "Nevada", "Nevada"],
        "year":[2001,2002,2003,2004,2005,2006],
        "pop":[1.5,1.7,3.6,2.4,2.9,3.2]}
frame = pd.DataFrame(data)
frame
```

```python
frame.head()
```

```python
pd.DataFrame(data, columns=["year","state"]) # 원하는 순서대로, 원하는 열만 지정
```

```python
frame2 = pd.DataFrame(data, columns=["year","state","pop","debt"]) # 딕셔너리에 없는 값은 Na
frame2
```

```python
frame2["state"]
```

```python
frame2.loc[1]
```

```python
frame2.iloc[2]
```

```python
frame2["debt"] = np.arange(6.) # 대입으로 열 전체 수정
frame2
```

```python
val = pd.Series([-1.2, -1.5, -1.7], index=['two', 'four', 'five'])
frame2['debt'] = val # 색인에 따라 값이 대입. 존재하지 않으면 결측치가 대입된다
frame2
```

```python
frame2["isohio"] = (frame2["state"] == "Ohio") # 불리언 열 추가
frame2
```

```python
del frame2["isohio"] # 열 삭제
frame2.columns
```

```python
# 중첩된 딕셔너리
population = {"ohio": {2000: 1.5, 2001:1.7, 2002:3.6},
              "Nevada": {2001:2.4, 2002:2.9}}
frame3 = pd.DataFrame(population)
frame3
```

```python
pd.DataFrame(population, index=[2001,2002,2003])
```

```python
frame3.index.name = "year"
frame3.columns.name = "state"
print(frame3)
```
