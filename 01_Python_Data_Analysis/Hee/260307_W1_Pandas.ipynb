```python
# 인덱스가 다른 배열의 연산. 서로 없는 값은 결측치로 바꿈/ DataFrame의 경우는 행과 열에 모두 결측치
a = pd.Series(np.arange(3), index=['a','b','c'])
b = pd.Series(np.arange(4), index=['a','b','d','e'])
a+b
```

```python
df1 = pd.DataFrame({"A":[1,2]})
df2 = pd.DataFrame({"b":[3,4]})
df1.add(df2, fill_value=0) # 이렇게 하면 없는 값을 0으로 해서 결측치로 변하는거 방지함 add, sub, div, mul, pow, floordiv(소수점 내림 연산 //)
                                                                                    # radd, rsub, rdiv ..... 뒤집어서 계산 1/df == df.rdiv(1)
```

```python
arr = np.arange(12.).reshape((3,4))
arr - arr[0] # 각 행에 대해서 한 번씩 수행 (브로드캐스팅)
```

```python
# DataFrame도 유사함
frame = pd.DataFrame(np.arange(12.).reshape((4,3)),
                     columns=['b','d','e'],
                     index=['Utah', 'Ohio', 'Texas', 'Oregon'])
frame
```

```python
series = frame.iloc[0]
frame - series # 아래 행으로 브로드캐스팅
```

```python
series2 = frame['d']

frame.sub(series2, axis=0) # 각 행에 대해 연산 수행
```

### 2.5 함수 적용과 매핑

```python
frame = pd.DataFrame(np.random.standard_normal((4,3)),
                     columns=['b','d','e'],
                     index=['Utah', 'Ohio', 'Texas', 'Oregon'])
frame
```

```python
np.abs(frame)
```

```python
def f1(x):
    return x.max() - x.min()

frame.apply(f1)
```

```python
frame.apply(f1, axis='columns') # 각 행에 대해서 한 번씩만 수행
```

```python
def f2(x):
    return pd.Series([x.min(), x.max()], index=['min','max'])
frame.apply(f2)
```

```python
def my_format(x):
    return f"{x:.2f}"
frame.applymap(my_format)
```
