```python
import pandas as pd
import numpy as np
float_data = pd.Series([1.2, -3.5, np.nan, 0])
float_data
```

```python
data = pd.DataFrame([[1., 6.5, 3.],
                     [np.nan, np.nan, np.nan],
                     [1., np.nan, np.nan],
                     [np.nan, 6.5, 3.]])
data
```

```python
data.dropna(how='all') # 행이 모두 nan인 행만 제거
```

```python
data[4] = np.nan
data
```

```python
data.dropna(axis='columns', how='all') # 모두 nan인 열 제거
```

```python
data = pd.DataFrame({"k1": ["one", "two"]*3 + ["two"],
                     "k2": [1,1,2,3,3,4,4]})
data
```

```python
data.duplicated() # 각 행이 중복인지
```

```python
data.drop_duplicates() # 6행 제거
```

```python
data['v1'] = range(7)
data
```
