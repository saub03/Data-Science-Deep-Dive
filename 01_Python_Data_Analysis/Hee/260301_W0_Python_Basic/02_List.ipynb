## 리스트 다루기

```python
a = [1, 2, 3]
a.append(4) # 마지막에 4 추가
a.sort() # 오름차순 정렬
a.reverse() # 역순 4 3 2 1
a.insert(0, 5) # 0 번째에 5 추가 5 4 3 2 1
a.remove(5) # 첫 번째로 나오는 5 삭제
a.pop() # 맨 오른쪽 요소 반환하고 삭제
a.extend([1, 0]) # 익스텐드
a.count(1) # 1의 개수
print(a.index(2))
```
