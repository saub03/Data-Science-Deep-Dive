# dict

```python
dic = {'name': '홍길동', 'phone': '010-9999-1234', 'birth': '1118'}
# 쌍 추가
dic['loc'] = 'poland'
print(dic)
```

```python
del dic['loc']
print(dic)
```

## dict 다루기
keys values items는 for 반복문에서 많이 사용하므로 익혀두기

```python
print(dic.keys()) # 키 리스트
print(dic.values()) # 값 리스트
print(dic.items()) # 아이템 리스트 키와 값을 튜플화

dic.clear()  # 모두 지우기
print(dic)
dic = {'name': '홍길동', 'phone': '010-9999-1234', 'birth': '1118'}

print(dic.get('name')) # 키로 값 얻기, dic['name'] 과 달리 존재하지 않는 키값이면 None 반환
print('name' in dic) # 키가 있는지 조사 -> bool
print(dic.pop("phone")) # 키의 값을 반환하고 쌍 삭제
print(dic)
```
