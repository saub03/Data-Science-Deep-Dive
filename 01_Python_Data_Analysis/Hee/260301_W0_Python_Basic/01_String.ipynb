## 문자열 포멧팅 -> 딱히 쓸모 없으니까 넘어가셈


```python
"I eat %d apples" % 3
"I eat %s apples" % "three"
"%d is %s" % (3, "three")
```

```python
print("I eat {0}, apples".format("5"))
print("I eat {0}, {1}".format("5", "apples"))
```

```python
"%10s" % 'hi' #________hi 10칸 공백중 오른쪽 정렬
"%-10s" % 'h1' #hi__________ 

"{0:<10}".format("hi") # 왼쪽 정렬
"{0:>10}".format("hi") # 오른쪽 정렬
"{0:^10}".format("hi") # 가운데 정렬
```

## 문자열 변환 기법!

```python
print("hello".count('l'))
print("hello".find('l')) # 처음으로 나온 위치 반환, 없으면 -1
print("hello".index('l')) # 처음으로 나온 위치, 없으면 오류
print(", ".join('hello')) # 문자열 삽입
print(", ".join(['a', 'b', 'c', 'd'])) # 리스트나 튜플에도 사용 가능
print("hi".upper()) # 대문자 변환
print("HI".lower()) # 소문자 변환
print("   left   ".lstrip()) # 왼쪽 공백 삭제
print("   right   ".rstrip()) # 오른쪽 공백 삭제
print("   space   ".strip()) # 양쪽 공백 삭제
print("how to replace string".replace('replace', 'change')) # 문자열 바꾸기
print("how to replace string".split()) # 문자열 나누기 -> list
print("abcd".isalpha()) # 문자열이 알파벳으로만 이루어져있나? -> bool
print("123".isdigit()) # 문자열이 숫자로만 이루어져있나? -> bool
print("This string stats with T".startswith('T')) # 특정 문자(열)로 시작하는지? -> bool
print("This string stats with T".endswith('t')) # 특정 문자(열)로 끝나는지? -> bool

```
