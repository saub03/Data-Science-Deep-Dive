# for

```python
l1 = ['one', 'two', 'three']

for i in l1:
  print(i)
```

```python
for i in range(0, 10):
  print(i)
```

```python
for i, j in enumerate(l1):
  print(f"{i} {j}")
```

```python
l1 = ['가', '나', '다']
l2 = [12, 34, 56]
for name,num in zip(l1, l2):
  print(f"{name}{num}")
```
