# 함수

```python
def add_many(*args):
  result = 0
  for i in args:
    result += i
    
  return result

print(add_many(1, 2, 3, 4, 5))
```

```python
def add_many(mul,*args):
  result = 0
  if mul == 'add':
    for i in args:
      result += i
    return result
  elif mul == 'mul':
    result = 1
    for i in args:
      result *= i
    return result
  
print(add_many('add', 1,2,3,4,5))
print(add_many('mul', 1,2,3,4,5))
```

```python
def print_kwargs(**kwargs):
  print(kwargs)
  
print_kwargs(name = 'foo', age = '3')
```

```python
def mul_arg():
  return 1, 2

a = mul_arg()
print(a) #-> tuple
```

## 람다를 통해 한 줄로 함수를 선언할 수 있음

```python
add = lambda a, b: a+b
result = add(1,2)
print(result)
```

```python
def add(a, b):
    """
    두 숫자를 더하는 함수

    Parameters:
    a (int, float): 첫 번째 숫자
    b (int, float): 두 번째 숫자

    Returns:
    int, float: 두 숫자의 합
    """
    return a + b

# 독스트링 확인하기
print(add.__doc__)
```
