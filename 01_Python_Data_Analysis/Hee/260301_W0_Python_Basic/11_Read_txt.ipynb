# open
보통 데이터 가공을 할 때 SQL과 numpy,pandas로 처리를 함. 데이터를 불러올 때 보통 pd.read_csv 같은 고수준 도구를 사용하지만 파이썬 자체에서 파일을 다루는지 이해하는것도 중요함.
파일을 읽고 쓰기 위해서는 내장 함수인 open을 이용함

```python
path = "text.txt"

f = open(path, encoding="utf-8")
for line in f:
    print(line)
```

파일 줄에는 EOL 문자가 있으므로 rstrip 해줌

```python
lines = [x.rstrip() for x in open(path, encoding="utf-8")]
lines
```

open을 사용하면 명시적으로 닫아줘야 메모리를 반환함

```python
f.close()
```

with 을 사용하면 with 블록이 끝나는 시점에서 파일 핸들 f가 자동으로 close 됨

```python
with open(path, encoding="utf-8") as f:
    lines = [x.rstrip() for x in f]
    
print(lines)
```
