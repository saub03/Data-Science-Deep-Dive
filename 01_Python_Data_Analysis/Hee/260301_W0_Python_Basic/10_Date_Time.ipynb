# datetime 라이브러리
시계열 분석 시 사용하니까 잘 알아두어야함

datetime 모듈은 datetime, date, time 자료형을 지원한다

```python
from datetime import datetime, date, time
```

```python
dt = datetime(2026, 3, 10, 20, 30, 12) # 2026-3-10 20:30 12초
```

date와 time 메서드로 날짜와 시간 추출

```python
print(dt.date())
print(dt.time())
```

strftime 메서드

```python
dt.strftime("%Y-%m-%d %H:%M:%S") # 문자열로 포맷
```

strptime 메서드

```python
datetime.strptime("20260301", "%Y%m%d") # 문자열을 datetime 객체로 변환
```

필드 치환

```python
dt_replaced = dt.replace(minute=0, second=0) # 이 메서드는 항상 새로운 객체를 반환한다 (딥카피)
dt_replaced
```
