```python
def isiterable(obj):
    try:
        iter(obj)
        return True
    except TypeError:
        return False
    

a = [1, 2, 3]
b = "TestString"
c = 5

print(isiterable(a))
print(isiterable(b))
print(isiterable(c))
```
