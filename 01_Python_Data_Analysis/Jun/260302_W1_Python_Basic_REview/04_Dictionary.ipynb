## 4. Dict(Dictionary)

### 4.1. 기초

```python
dic = {'name': '전준황', 'age': 18, 'birth':'0601'}
print(f"Dictionary는 C, C++의 map과 비슷하다. 해시맵이라고 하는 게 더 정확할 지도\n{dic}")
del dic['birth']
print(f"\n때문에, del을 할 때 key 값을 이용해 지울 수 있다. 개인정보인 생일 같은 거?\n{dic}")
print(f"\n하지만, del을 할 때 key 값이 아닌 value 값을 이용해 지울 순 없다.")
```

### 4.2. Method 활용

```python
print(f"keys() 매서드를 활용하면 dic의 key 값들을 리스트 형태로 출력한다.\n{dic.keys()}")
print(f"value() 매서드는 반대로 value 값들을 리스트 형태로 출력한다.\n{dic.values()}")
print(f"items() 매서드는 key와 value를 튜플 형태로 묶어서 출력해준다.\n{dic.items()}")
print(f"get() 매서드는 key 값을 입력하여 해당하는 value 값을 얻는다. 없을 경우에는 None을 반환함.\nResult: {dic.get('name')}")
print("in 연산자를 통해 해당하는 key가 dic에 있는지 bool을 통해 확인할 수 있다")
print('name' in dic)
print(f"pop() 매서드는 key 값을 입력하여 해당하는 value를 반환하고 삭제한다.\nDeleted: {dic.pop('age')}")
print(f"최종결과: {dic}")

dic.clear()
print(f"clear() 매서드는 dic을 전부 비워준다.\n{dic}")
# 만약 위 매서드 예제도 dic이 비어있다면 이전 코드를 실행하길 바랍니다.
```
