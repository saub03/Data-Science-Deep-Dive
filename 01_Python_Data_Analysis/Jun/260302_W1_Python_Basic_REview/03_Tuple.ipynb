## 3. Tuple

### 3.1. 기본

```python
t1 = (1, 2, 3)
t2 = 1, 2, 3
t3 = (1,)

print(f"{t1}, {t2}, {t3}는 모두 {type(t1)}, {type(t2)}, {type(t3)} 입니다.")
print("C언어의 const와 비슷한 속성이지만 완전히 같지는 않으며")
print("튜플이 가르키는 객체가 mutable 인지 immutable 인지에 따라 갈림\n")
t4 = (1, 2, [9, 0])
print(f"{t4} 라는 튜플이 있다고 가정하고 2번째 인덱스인 리스트를 건드리면")
t4[2].append(4)
print(f"{t4} 처럼 수정이 가능하다\n")
print("그리고 index나 count 매소드를 사용할 수 있다")
print(f"{t4.index(2)}: index(2), {t4.count(1)}: count(1)")
```
