### [!] 10. 그룹화 및 집계 (GroupBy)

```python
df_clean.groupby("species")
df_clean
df_clean.groupby('species')['body_mass_g'].agg(["mean"])
# df_clean.(의 데이터를)groupby(다음으로 group을 함)('species')['body_mass_g'].(의 데이터를)agg(["mean"])(으로 집계한다)
"""
df.clean 의 데이터의
.groupby 매서드인 groupby를 사용해서
('species')['body_mass_g'] 종의 g 무게를
.agg() 로 집계 하는데
['mean'] 중앙값으로 집계한다.
"""
```

```python
df_clean.groupby('species')['body_mass_g'].mean()
df_clean.groupby('species')['body_mass_g'].agg(['mean', 'count'])
# agg 안에 입력 되는 것들을 데이터프레임 형식으로 뽑아내줌
# mean은 중앙값만 짠 내주지만 agg() 는 넣는대로 보여줌
# 대괄호를 넣는 이유는, agg()가 여러 개를 받아오기 때문임
```

```python
df_clean[['island', 'species']].value_counts()
df_clean[['species', 'island']].value_counts()
# 펭귄 종까지 알아오라고 하진 않았으니까 그냥 출력하면 될듯
tempdf = df_clean['island'].value_counts()

for island, cnt in tempdf.items():
    print(f"{island:<10} 에는 {cnt:>5} 마리의 펭귄이 삽니다.")
```
