### 3. 결측지 개수 파악
각 컬럼별로 결측치(NaN)가 총 몇 개씩 있는지 확인하세요.

```python
colList = df.columns.tolist()

for i in colList:
    print(f"컬럼 {i:18}에는 {df[i].isna().sum()}개의 결측치가 있습니다.")
```

### 4. 결측치 처리 (행 삭제)
결측치가 하나라도 포함된 행을 모두 제거하고, 그 결과를 df_clean이라는 새로운 변수에 저장하세요. (이후 문제는 df_clean을 사용합니다.)

```python
df_copied = df.copy()
df_clean = df_copied.dropna()

# df_clean.info() 를 통해 결측치 삭제 되었는 지 확인
df_clean
```
