## 6. Copy

### 1. 배경지식

먼저 'Shallow Copy' 와 'Deep Copy'의 차이점을 알아야함.

Shallow Copy는 C언어의 포인터를 생각하면 좀 비슷하다.
예를 들어 (new) B list = A list 를 한다고 가정하자. 그럼 두 리스트는 서로 같은 값을 가지게 될 것이다.
그런데? B list의 값을 바꾸면 A list의 값도 같이 바뀌게 된다.
마치 C언어의 함수 매개변수 포인터를 보는 거 같지 않나?

파이썬은 메모리를 효율적으로 사용하는 것을 중요시 하기 때문에
이러한 copy 양상을 띄게 된다.

Deep Copy는 Shallow Copy와 다르게 독립적인 객체를 만들어낸다.
때문에 B list의 값을 바꾸더라도 A list의 값은 변하지 않는다.

이를 사용하려면 copy 라이브러리를 임포트 해야한다.

### 2. Copy 함수

```python
from copy import copy # copy 라이브러리에서 copy 함수를 가져옴.

print("ㅡㅡㅡㅡㅡShallow Copyㅡㅡㅡㅡㅡ")
a = list("전준황")
print(f"{a[0]} {a[1]} {a[2]}")
b = a
print(f"{b[0]} {b[1]} {b[2]}")
b[0] = '쨈'
print(f"{'\033[95m'}{a[0]}{'\033[0m'} {a[1]} {a[2]}")
print(f"{'\033[95m'}{b[0]}{'\033[0m'} {b[1]} {b[2]}")
print("ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ")

print("ㅡㅡㅡㅡㅡㅡDeep Copyㅡㅡㅡㅡㅡㅡ")
a = list("전준황")
print(f"{a[0]} {a[1]} {a[2]}")
b = copy(a)
print(f"{b[0]} {b[1]} {b[2]}")
b[0] = '쨈'
print(f"{'\033[95m'}{a[0]}{'\033[0m'} {a[1]} {a[2]}")
print(f"{'\033[95m'}{b[0]}{'\033[0m'} {b[1]} {b[2]}")
print("ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ")
```

```python
a, b = (3, 4)
print(f"a: {a}, b: {b}")
a, b = b, a
print(f"a: {a}, b: {b}")
```
