## 5. Set

### 5.1. 기초

```python
eg = set([1, 2, 2, 3])
print(f"set 자료형은 중복된 값을 허용하지 않는다.\n{eg}")
s1 = set([1, 2, 3])
s2 = set([3, 4, 5])

sR = s1&s2
print(f"교집합: {sR}")
sR = s1|s2
print(f"합집합: {sR}")
sR = s1-s2
print(f"차집합: {sR}")
```

### 5.2. Methods 활용

```python
s1.remove(1)
print(f"remove() 메서드는 set 자료형에서 특정 값을 제거한다.\n{s1}")
s1.add(1)
print(f"add() 메서드는 set 자료형에서 특정 값을 추가한다.\n{s1}")
s1.update([4, 5])
print(f"update() 메서드는 set 에서 여러 값을 추가할 수 있다. for문 할 필요 없음 ㄷㄷ\n{s1}")
print(f"discord() 매서드는 remove()와 같이 제거하는 매서드이지만, 존재하지 않아도 오류를 띄우지 않는다.\n{s1.discard(6)}")
print(f"clear() 매서드는 set을 전부 비워줌.\n{s1.clear()}")
```
