## 7. Control Flow (제어흐름)

### 1. 기본

```python
for i in range(0, 9):
    print(i)
    if i == 5:
        print("이런 식으로 조건문을 둘 수 있다.")
print("ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ")
count = 0
found = False
while not found:
    count += 1
    if count == 5:
        found = True
        print("찾았다!!")
```

### 2. 심?화

```python
nameList = ["Junhwang", "Heewoo", "Gildong"]
nameinput = input("찾는 사람의 이름을 입력: ")
for name in nameList:
    if name == nameinput:
        print(f"{name}을 발견했다!!")
        nameList.remove(name)
    else:
        print(f"{name}은 우리가 찾는 사람이 아닙니다.")
```

```python
peopleList = {};

count = 1
while count != 0:
    index = count
    name = input("이름: ")
    if (name == "0"): break
    age = int(input("나이: "))

    addInfo = {"index": index, "name": name, "age": age}
    peopleList[index] = addInfo
    
    count += 1


print(peopleList)
```
