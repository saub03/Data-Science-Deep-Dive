### 1.5. Method 활용

```python
print("This is example-desu.".count('i'), "| .count(): string 내에 세고자 하는 단어의 개수를 보여줌")
print("TASTE THE RAINBOW".find("R"), "| .find(): string 내에 찾고자 하는 것이 최초로 나타나는 인덱스를 출력함\n없으면 -1")
print("Tomorrow 수강변경기간 jet-tai-ni success 이카나캬".index("o"), "| .index(): find()랑 다르게 인덱스 못 찾으면 오류를 띄움")
print("★".join("STAR"), "| .join(): 괄호 안 string 사이사이에 문자 넣어줌")
print("☆".join(['와', '신', '기', '해']), "| 놀랍게도 리스트, 튜플에도 사용가능하다.")
print("everything is lowercase!!!".upper(), "| .upper(): 소문자를 전부 대문자로!!!")
print("EVERYTHING IS UPPERCASE!!!".lower(), "| .lower(): 대문자를 전부 소문자로!!!")
print("   CUT LEFT SPACE!!    ".lstrip(), "| .lstrip(): l은 left, strip은 벗겨내다, 떼어내다.")
print("   CUT RIGHT SPACE!!    ".rstrip(), "| .rstrip(): 위와 같음. 각자의 가장 끝을 잘라줌.")
print("아 이거 마음에 안 드는데?".replace("이거", "저거"), "| .replace(A, B): A를 B로 바꿔준다~")
print("야 이거 분류 좀 해라.".split(), "| .split(): 띄어쓰기 구분해서 리스트로 나눔!")
print("alphaisthatyou".isalpha())
print("alpha is that you?".isalpha(), "| .isalpha(): >>알파벳<< 으로만 이루어졌는지")
print("1234567".isdigit(), "| .isdigit(): 위랑 같음")
print("이것이 시작인가요?".startswith("이것"), "| .startswith(): 이걸로 시작하나요?")
print("마지막이 이것으로 끝나나요?".endswith("마"), "| .endswith(): 이것으로 끝나나요?")
```

### 2.3. Method 활용

```python
example = [3, 1, 10]
example.append(4) # 마지막 인덱스에 추가함
print("{:10s} {}".format("append:", example))
example.sort() # 오름차순으로 정렬함 O(n log n)
print("{:10s} {}".format("sort:", example))
example.reverse() # 역순으로
print("{:10s} {}".format("reverse:", example))
example.insert(0, 21) # 해당하는 인덱스에 요소 추가
print("{:10s} {}".format("insert:", example))
example.remove(4) # 리스트 내에 삭제. 전부 삭제가 아닌, 최초 발견 요소를 삭제함
print("{:10s} {}".format("remove:", example))
example.pop() # 맨 오른쪽 삭제 C++의 vector pop_back ..뭐 그런 느낌?
print("{:10s} {}".format("pop:", example))
example.extend([1, 9]) # iterable 을 펼쳐서 추가해줌
print("{:10s} {}".format("extend 1:", example))
example.extend([[1, 2], [3, 10]]) # iterable은 한 번만 펼침.
print("{:10s} {}".format("extend 2:", example))
print("{:10s}개 {}".format("count:", example.count(1))) # 몇 개 있는 지. 깊게 관찰하지 않기 때문에 2개가 아니라 1개 출력함
```
