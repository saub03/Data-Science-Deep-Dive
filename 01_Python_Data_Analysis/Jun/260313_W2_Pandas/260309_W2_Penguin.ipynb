```python
df_fill = df.copy()
df_fill['bill_length_mm'] = df_fill['bill_length_mm'].fillna(df['bill_length_mm'].mean())
df_fill['sex'] = df_fill['sex'].fillna(df_fill['sex'].mode().iloc[0])
# mode()는 N 이라는 게 3번 나왔습니다 (N, 3) 튜플로
df_fill = df_fill.dropna()
df_fill.info()
```
